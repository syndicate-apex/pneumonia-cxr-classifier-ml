# Data Balancing for Pneumonia Dataset
This notebook analyzes and balances the class imbalance in the chest X-ray dataset

## 1. Check Current Imbalance

```python
import os
import shutil
from collections import Counter
import numpy as np

# Count files in each class
data_dir = "chest_xray/train"

normal_count = len(os.listdir(os.path.join(data_dir, "NORMAL")))
pneumonia_count = len(os.listdir(os.path.join(data_dir, "PNEUMONIA")))

print("=== Training Set Class Distribution ===")
print(f"NORMAL samples: {normal_count}")
print(f"PNEUMONIA samples: {pneumonia_count}")
print(f"Total: {normal_count + pneumonia_count}")
print(f"\nImbalance ratio: {pneumonia_count / normal_count:.2f}x more pneumonia samples")
print(f"NORMAL percentage: {100 * normal_count / (normal_count + pneumonia_count):.1f}%")
print(f"PNEUMONIA percentage: {100 * pneumonia_count / (normal_count + pneumonia_count):.1f}%")
```

## 2. Balance Dataset - Simple Duplication Method
This duplicates NORMAL samples to match PNEUMONIA count

```python
import random

normal_dir = os.path.join(data_dir, "NORMAL")
pneumonia_dir = os.path.join(data_dir, "PNEUMONIA")

normal_files = os.listdir(normal_dir)
pneumonia_files = os.listdir(pneumonia_dir)

# Calculate how many duplicates we need
duplicates_needed = pneumonia_count - normal_count

if duplicates_needed > 0:
    print(f"Adding {duplicates_needed} duplicate NORMAL samples...")
    
    # Randomly select files to duplicate
    files_to_duplicate = random.choices(normal_files, k=duplicates_needed)
    
    for i, file in enumerate(files_to_duplicate):
        src = os.path.join(normal_dir, file)
        # Create unique name for duplicate
        name, ext = os.path.splitext(file)
        dst = os.path.join(normal_dir, f"{name}_dup_{i}{ext}")
        shutil.copy2(src, dst)
    
    print(f"✓ Balanced! New NORMAL count: {len(os.listdir(normal_dir))}")
else:
    print("Dataset is already balanced or PNEUMONIA has fewer samples.")
```

## 3. Verify New Balance

```python
# Recount
normal_count_new = len(os.listdir(os.path.join(data_dir, "NORMAL")))
pneumonia_count_new = len(os.listdir(os.path.join(data_dir, "PNEUMONIA")))

print("=== Updated Training Set ===")
print(f"NORMAL samples: {normal_count_new}")
print(f"PNEUMONIA samples: {pneumonia_count_new}")
print(f"Total: {normal_count_new + pneumonia_count_new}")
print(f"\nNORMAL: {100 * normal_count_new / (normal_count_new + pneumonia_count_new):.1f}%")
print(f"PNEUMONIA: {100 * pneumonia_count_new / (normal_count_new + pneumonia_count_new):.1f}%")
print(f"\n✓ Dataset is now balanced!")
```
